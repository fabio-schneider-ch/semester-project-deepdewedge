```python
from pathlib import Path
import os
import pandas as pd
import matplotlib.pyplot as plt

metrics_path = Path("../../results/subexp02")

print("Current working directory:")
print(os.getcwd())

angles = [40, 50, 60]

metrics_paths = {}

for angle in angles:
    metrics_paths[angle] = {
        "angle": angle,
        "path": metrics_path / f"metrics_mw{angle}.csv",
    }

#print(metrics_paths)
```

```python
metrics = {}

for angle in angles:
    path = metrics_paths[angle]["path"]
    metrics[angle] = pd.read_csv(path)
    print(f"mw{angle}:")
    print(metrics[angle].columns)
    print(metrics[angle].head())
```

```python
plt.figure(figsize=(8, 5))

for angle in [40, 50, 60]:
    df = metrics[angle] #choosing a dataframe from the dict
    fit = df[df["fitting_loss"].notna()] #Ingoring NaN Values and takin val_loss column

    #Plot 
    plt.plot(
        fit["epoch"],
        fit["fitting_loss"],
        label=f"trained {angle}°",
        marker = 'x',
        markersize = 5,
        linewidth=0.8
    )

plt.xlabel("Epoch")
plt.ylabel("Fitting loss")
plt.title("Sub-Exp 2: fitting loss comparison")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

os.makedirs("../../figures/subexp02", exist_ok=True)
save_path = "../../figures/subexp02/fitting_loss_comparison_train_mw40_50_60.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")

plt.show()

print("Saved to:", os.path.abspath(save_path))
```

```python
plt.figure(figsize=(8, 5))

for angle in [40, 50, 60]:
    df = metrics[angle] #choosing a dataframe from the dict
    val = df[df["val_loss"].notna()] #Ingoring NaN Values and takin val_loss column

    #Plot
    plt.plot(
        val["epoch"],
        val["val_loss"],
        label=f"trained {angle}°",
        linewidth=1.2,
        marker="x",
        markersize=4
    )

plt.xlabel("Epoch")
plt.ylabel("Validation loss")
plt.title("Sub-Exp 2: validation loss comparison")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

os.makedirs("../../figures/subexp02", exist_ok=True)
save_path = "../../figures/subexp02/validation_loss_comparison_train_mw40_50_60.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")

plt.show()

print("Saved to:", os.path.abspath(save_path))
```

```python
plt.figure(figsize=(8, 5))

for angle in [40, 50, 60]:
    df = metrics_1000[angle] #choosing a dataframe from the dict
    val_1000 = df[df["val_loss"].notna()] #Ingoring NaN Values and takin val_loss column

    #Plot
    plt.plot(
        val_1000["epoch"],
        val_1000["val_loss"],
        label=f"trained {angle}°",
        linewidth=1.2,
        marker="x",
        markersize=4
    )
    best_idx = val_1000["val_loss"].idxmin()
    best_epoch = val_1000.loc[best_idx, "epoch"]
    best_val = val_1000.loc[best_idx, "val_loss"]
    
    plt.scatter(best_epoch, best_val, s=40, marker="o")
    plt.axvline(best_epoch, linestyle="--", alpha=0.4)
    print("Val_loss: ",best_val, "at epoch:", best_epoch)

plt.xlabel("Epoch")
plt.ylabel("Validation loss")
plt.title("Sub-Exp 2: validation loss comparison")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

os.makedirs("../../figures/subexp02", exist_ok=True)
save_path = "../../figures/subexp02/validation_loss_comparison_train_mw40_50_60_1000.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")

plt.show()

print("Saved to:", os.path.abspath(save_path))
```
