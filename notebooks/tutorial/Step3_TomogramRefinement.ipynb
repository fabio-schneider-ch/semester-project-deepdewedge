
## Refining the Tutorial Data

The `ddw refine-tomogram` produces the final denoised and missing-wedge-corrected tomogram(s) using a model checkpoint produced during model fitting. You can determine which checkpoint to use by setting the `model_checkpoint_file` parameter in the `config.yaml` file. You can find checkpoint files (`.ckpt` extension) in `./tutorial_project/logs/version_0/checkpoints`. This directory contains three sub-directories:

- `epoch/`: Checkpoints periodically created after a fixed number of epochs
- `fitting_loss/`: Checkpoints created when the model achieved lowest fitting losses 
- `val_loss/`: Checkpoints created when the model achieved the lowest validation losses

The frequnency and criteria for the creation of these checkpoints are arguments of `ddw fit-model`; see the output of `ddw fit-model --help` for details.  

Provided that there was no overfitting, the checkpoint corresponding to the highest epoch should give good results, but you can also try the ones with the lowest validation or fitting losses.

The following command uses a model checkpoint contained in `./tutorial_data` which was fitted by us. If you want to use a checkpoint from your own `ddw fit-model` run, change the `model_checkpoint_file` to, e.g., `./tutorial_project/logs/version_0/checkpoints/epoch/epoch=999.ckpt`.

Model checkpoints can be found in the logging directory used during `ddw fit-model`, which is typically `{project_dir}/logs`. In this concrete case, the `tutorial_project/logs` directory contains model checkpoints for various epochs, as well as those achieving the lowest validation and fitting losses. The frequency and number of checkpoints created is controlled by arguments during model fitting, see `ddw fit-model --help` for more information. 

````python
import os
import numpy as np
import matplotlib.pyplot as plt
import torch

from ddw.utils.mrctools import load_mrc_data

# Load tomograms
tomo_ddw = load_mrc_data("./tutorial_project/refined_tomograms/tomo_even_frames+tomo_odd_frames_refined.rec")
tomo_fbp = load_mrc_data("./tutorial_data/tomo_all_frames.rec")

# Convert to numpy if needed
if isinstance(tomo_ddw, torch.Tensor):
    tomo_ddw = tomo_ddw.cpu().numpy()
if isinstance(tomo_fbp, torch.Tensor):
    tomo_fbp = tomo_fbp.cpu().numpy()

# Middle slices
z_mid = tomo_fbp.shape[0] // 2
y_mid = tomo_fbp.shape[1] // 2

# XY slices (constant z)
fbp_xy = tomo_fbp[z_mid, :, :]
ddw_xy = tomo_ddw[z_mid, :, :]

# XZ slices (constant y)
fbp_xz = tomo_fbp[:, y_mid, :]
ddw_xz = tomo_ddw[:, y_mid, :]

# Common display range
all_std = max(tomo_fbp.std(), tomo_ddw.std())
vmin = -3 * all_std
vmax =  3 * all_std

# Create figure
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].imshow(fbp_xy, cmap="gray", vmin=vmin, vmax=vmax)
axes[0, 0].set_title("FBP XY")
axes[0, 0].axis("off")

axes[0, 1].imshow(ddw_xy, cmap="gray", vmin=vmin, vmax=vmax)
axes[0, 1].set_title("DeepDeWedge XY")
axes[0, 1].axis("off")

axes[1, 0].imshow(fbp_xz, cmap="gray", vmin=vmin, vmax=vmax)
axes[1, 0].set_title("FBP XZ")
axes[1, 0].axis("off")

axes[1, 1].imshow(ddw_xz, cmap="gray", vmin=vmin, vmax=vmax)
axes[1, 1].set_title("DeepDeWedge XZ")
axes[1, 1].axis("off")

plt.tight_layout()

# Save figure
os.makedirs("../../figures", exist_ok=True)
save_path = "../../figures/day5_input_vs_refined_xy_xz.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")
print(f"Saved figure to: {save_path}")
````
