```python
import os
import pandas as pd
import matplotlib.pyplot as plt

metrics_path = "../../results/training_runs/fast_200_epochs/metrics.csv"

print("Current working directory:")
print(os.getcwd())

print("\nMetrics file exists:")
print(os.path.exists(metrics_path))

metrics = pd.read_csv(metrics_path)

print("\nColumns:")
print(metrics.columns)

metrics.head()
```

```python
metrics[["epoch", "step", "fitting_loss", "val_loss"]].dropna(how="all").tail(20)
```

```python
# Extract non-empty loss values
fit = metrics[metrics["fitting_loss"].notna()]
val = metrics[metrics["val_loss"].notna()]

plt.figure(figsize=(8, 5))

plt.plot(
    fit["epoch"],
    fit["fitting_loss"],
    label="fitting loss",
    linestyle="--", linewidth=1.0, marker="x", markersize=4
)

plt.plot(
    val["epoch"],
    val["val_loss"],
    label="validation loss",
    linewidth=1.0, marker="x", markersize =5
)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("DeepDeWedge fast training loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

os.makedirs("../../figures", exist_ok=True)
save_path = "../../figures/day4_training_loss_curve.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")

plt.show()

print("Saved to:", os.path.abspath(save_path))
print("File exists:", os.path.exists(save_path))
```
